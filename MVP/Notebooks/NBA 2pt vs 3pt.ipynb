```python
#winner 3pt attempt by decades
eighty_winner=winner_filter.loc[(winner_filter["Year"]>=1980) & (winner_filter["Year"]<=1990)]
ninety_winner=winner_filter.loc[(winner_filter["Year"]>1990) & (winner_filter["Year"]<=2000)]
thousand_winner=winner_filter.loc[(winner_filter["Year"]>2000) & (winner_filter["Year"]<=2010)]
ten_winner=winner_filter.loc[(winner_filter["Year"]>2010) & (winner_filter["Year"]<=2018)]
eighty_winner_attempt=eighty_winner["TPA"].sum()
ninety_winner_attempt=ninety_winner["TPA"].sum()
thousand_winner_attempt=thousand_winner["TPA"].sum()
ten_winner_attempt=ten_winner["TPA"].sum()
```

```python
#loser 3pt attempt by decades
eighty_loser=loser_filter.loc[(loser_filter["Year"]>=1980) & (loser_filter["Year"]<=1990)]
ninety_loser=loser_filter.loc[(loser_filter["Year"]>1990) & (loser_filter["Year"]<=2000)]
thousand_loser=loser_filter.loc[(loser_filter["Year"]>2000) & (loser_filter["Year"]<=2010)]
ten_loser=loser_filter.loc[(loser_filter["Year"]>2010) & (loser_filter["Year"]<=2018)]
eighty_loser_attempt=eighty_loser["TPA"].sum()
ninety_loser_attempt=ninety_loser["TPA"].sum()
thousand_loser_attempt=thousand_loser["TPA"].sum()
ten_loser_attempt=ten_loser["TPA"].sum()
```

```python
#1980~1990 mean
eighty_attempt_mean=(eighty_winner_attempt+eighty_loser_attempt)/2
eighty_attempt_mean
```

```python
#1990~2000 mean
ninety_attempt_mean=(ninety_winner_attempt+ninety_loser_attempt)/2
ninety_attempt_mean
```

```python
#2000~2010 mean
thousand_attempt_mean=(thousand_winner_attempt+thousand_loser_attempt)/2
thousand_attempt_mean
```

```python
#2010~2018 mean
ten_attempt_mean=(ten_winner_attempt+ten_loser_attempt)/2
ten_attempt_mean
```

```python
#3pt attempt dataframe
three_attempt_decade_mean=[eighty_attempt_mean,ninety_attempt_mean,thousand_attempt_mean,ten_attempt_mean]
three_attempt_decade_mean

decade=["1980~1990","1991~2000","2001~2010","2011~2018"]

decade_data_attempt=pd.DataFrame({"Year":decade,"three attempt mean":three_attempt_decade_mean,})
decade_data_attempt_new=decade_data_attempt.set_index('Year')

decade_data_attempt_new.plot(kind='bar')
plt.xlabel("Decade")
plt.ylabel("Total 3pt Attempts")
plt.legend(["Total 3pt Attempts"])
plt.xticks(rotation=0)
plt.title("3 pts Attempts Through Decades")
plt.savefig("3 pts Attempts Through Decades")
plt.tight_layout()
plt.show()
```

```python
#winner 3pt percentage by decades
eighty_winner=winner_filter.loc[(winner_filter["Year"]>=1980) & (winner_filter["Year"]<=1990)]
ninety_winner=winner_filter.loc[(winner_filter["Year"]>1990) & (winner_filter["Year"]<=2000)]
thousand_winner=winner_filter.loc[(winner_filter["Year"]>2000) & (winner_filter["Year"]<=2010)]
ten_winner=winner_filter.loc[(winner_filter["Year"]>2010) & (winner_filter["Year"]<=2018)]
eighty_winner_percentage=eighty_winner["TPP"].mean()
ninety_winner_percentage=ninety_winner["TPP"].mean()
thousand_winner_percentage=thousand_winner["TPP"].mean()
ten_winner_percentage=ten_winner["TPP"].mean()
```

```python
#loser 3pt percentage by decades
eighty_loser=loser_filter.loc[(loser_filter["Year"]>=1980) & (loser_filter["Year"]<=1990)]
ninety_loser=loser_filter.loc[(loser_filter["Year"]>1990) & (loser_filter["Year"]<=2000)]
thousand_loser=loser_filter.loc[(loser_filter["Year"]>2000) & (loser_filter["Year"]<=2010)]
ten_loser=loser_filter.loc[(loser_filter["Year"]>2010) & (loser_filter["Year"]<=2018)]
eighty_loser_percentage=eighty_loser["TPP"].mean()
ninety_loser_percentage=ninety_loser["TPP"].mean()
thousand_loser_percentage=thousand_loser["TPP"].mean()
ten_loser_percentage=ten_loser["TPP"].mean()
```

```python
#3pt percentage mean
eighty_percentage_mean=(eighty_winner_percentage+eighty_loser_percentage)/2
ninety_percentage_mean=(ninety_winner_percentage+ninety_loser_percentage)/2
thousand_percentage_mean=(thousand_winner_percentage+thousand_loser_percentage)/2
ten_percentage_mean=(ten_winner_percentage+ten_loser_percentage)/2
```

```python
#3pt percentage dataframe
three_percentage_decade_mean=[eighty_percentage_mean,ninety_percentage_mean,thousand_percentage_mean,ten_percentage_mean]
three_percentage_decade_mean

decade=["1980~1990","1991~2000","2001~2010","2011~2018"]

decade_percentage_data=pd.DataFrame({"Year":decade,"three percentage mean":three_percentage_decade_mean,})
decade_percentage_data_new=decade_percentage_data.set_index("Year")

decade_percentage_data_new.plot(kind='bar')


plt.xlabel("Decade")
plt.ylabel("3pt percentage")
plt.legend(["3pt percentage"])
plt.xticks(rotation=0)
plt.title("3 pts Percentage Through Decades")
plt.savefig("3 pts Percentage Through Decades")
plt.tight_layout()
plt.show()
```

```python
#3 pts attempt% through years
two_sum=winner_filter.groupby('Year')["FGA"].sum()
three_sum=winner_filter.groupby('Year')["TPA"].sum()
two_sum2=loser_filter.groupby('Year')["FGA"].sum()
three_sum2=loser_filter.groupby('Year')["TPA"].sum()
total_sum=two_sum+three_sum+two_sum2+three_sum2
two_percent=((two_sum+two_sum2)/total_sum)*100
three_percent=((three_sum+three_sum2)/total_sum)*100
year=np.arange(1980,2019)
plt.bar(year,two_percent,label="2 pt")
plt.bar(year,three_percent,label="3 pt")
plt.title("2 pts & 3 pts Attempt% Through Years")
plt.xlabel("Year")
plt.ylabel("Percentage")
plt.legend()
plt.savefig("2 pts & 3 pts Attempt% Through Years")
plt.show()
```

```python
#attempt scatter plot for each year
attempt=winner_filter[["Year","TPA"]]
year=np.arange(1980,2019)
TPA=attempt.groupby('Year')["TPA"].sum()

attempt2=loser_filter[["Year","TPA"]]
TPA2=attempt2.groupby('Year')["TPA"].sum()

plt.scatter(year,TPA+TPA2)
plt.plot(year,TPA+TPA2)
#curve_fit()

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(year, TPA+TPA2)
fit = slope * year + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.plot(year,fit,"--",color="red")
plt.annotate(line_eq,(1995,50),fontsize=15,color="red")
plt.title("3pt Attempt For Each Year")
plt.xlabel("Year")
plt.ylabel("Total 3pt attempt")
plt.savefig("3pt Attempt For Each Year")
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
winner=pd.read_csv("championsdata.csv")
winner_filter=winner[["Year","Team","Game","Win","TP","TPA","TPP"]]
winner_filter
```

```python
loser=pd.read_csv("runnerupsdata.csv")
loser_filter=loser[["Year","Team","Game","Win","TP","TPA","TPP"]]
loser_filter
```

```python
winner_3pt_percentage=winner_filter["TPP"].mean()
winner_3pt_percentage
```

```python
loser_3pt_percentage=loser_filter["TPP"].mean()
loser_3pt_percentage
```

```python
TPP_year_winner=winner_filter.groupby('Year')["TPP"].mean()
TPP_year_winner
```

```python
TPP_year_loser=loser_filter.groupby('Year')["TPP"].mean()
TPP_year_loser
```

```python
graph=TPP_year_winner.plot(label='Winner')
graph.set_title("3pt Percentage Through Years For Winner")
graph.set_xlabel('Year')
graph.set_ylabel('3pt Percentage')

graph=TPP_year_loser.plot(label='Loser')
graph.set_title("3pt Percentage Through Years For Loser")
graph.set_xlabel('Year')
graph.set_ylabel('3pt Percentage')
plt.legend()
plt.show()
```

```python
three_attempt_winner=winner_filter.groupby('Year')["TPA"].mean()
graph=three_attempt_winner.plot()
graph.set_title("3pt attempt Through Years For Winner")
graph.set_xlabel('Year')
graph.set_ylabel('3pt attempt')
plt.show()
```

```python
three_attempt_loser=loser_filter.groupby('Year')["TPA"].mean()
graph=three_attempt_loser.plot()
graph.set_title("3pt attempt Through Years For Loser")
graph.set_xlabel('Year')
graph.set_ylabel('3pt attempt')
plt.show()
```

```python
three_made_winner=winner_filter.groupby('Year')["TP"].mean()
graph=three_made_winner.plot()
graph.set_title("3pt made Through Years For Winner")
graph.set_xlabel('Year')
graph.set_ylabel('3pt made')
plt.show()
```

```python
three_made_loser=loser_filter.groupby('Year')["TP"].mean()
graph=three_made_loser.plot()
graph.set_title("3pt made Through Years For Loser")
graph.set_xlabel('Year')
graph.set_ylabel('3pt made')
plt.show()
```

```python
three_percentage_winner=winner_filter.groupby('Year')["TPP"].mean()
graph=three_percentage_winner.plot()
graph.set_title("3pt percentage Through Years For Winner")
graph.set_xlabel('Year')
graph.set_ylabel('3pt percentage')
plt.show()
```

```python
three_percentage_loser=loser_filter.groupby('Year')["TPP"].mean()
graph=three_percentage_loser.plot()
graph.set_title("3pt percentage Through Years For Loser")
graph.set_xlabel('Year')
graph.set_ylabel('3pt percentage')
plt.show()
```

```python
eighty_winner=winner_filter.loc[(winner_filter["Year"]>=1980) & (winner_filter["Year"]<=1990)]
ninety_winner=winner_filter.loc[(winner_filter["Year"]>1990) & (winner_filter["Year"]<=2000)]
thousand_winner=winner_filter.loc[(winner_filter["Year"]>2000) & (winner_filter["Year"]<=2010)]
ten_winner=winner_filter.loc[(winner_filter["Year"]>2010) & (winner_filter["Year"]<=2018)]
eighty_winner_attempt=eighty_winner["TPA"].mean()
ninety_winner_attempt=ninety_winner["TPA"].mean()
thousand_winner_attempt=thousand_winner["TPA"].mean()
ten_winner_attempt=ten_winner["TPA"].mean()
```

```python
eighty_loser=loser_filter.loc[(loser_filter["Year"]>=1980) & (loser_filter["Year"]<=1990)]
ninety_loser=loser_filter.loc[(loser_filter["Year"]>1990) & (loser_filter["Year"]<=2000)]
thousand_loser=loser_filter.loc[(loser_filter["Year"]>2000) & (loser_filter["Year"]<=2010)]
ten_loser=loser_filter.loc[(loser_filter["Year"]>2010) & (loser_filter["Year"]<=2018)]
eighty_loser_attempt=eighty_loser["TPA"].mean()
ninety_loser_attempt=ninety_loser["TPA"].mean()
thousand_loser_attempt=thousand_loser["TPA"].mean()
ten_loser_attempt=ten_loser["TPA"].mean()
```

```python
eighty_attempt_mean=(eighty_winner_attempt+eighty_loser_attempt)/2
eighty_attempt_mean
```

```python
ninety_attempt_mean=(ninety_winner_attempt+ninety_loser_attempt)/2
ninety_attempt_mean
```

```python
thousand_attempt_mean=(thousand_winner_attempt+thousand_loser_attempt)/2
thousand_attempt_mean
```

```python
ten_attempt_mean=(ten_winner_attempt+ten_loser_attempt)/2
ten_attempt_mean
```

```python
three_attempt_decade_mean=[eighty_attempt_mean,ninety_attempt_mean,thousand_attempt_mean,ten_attempt_mean]
three_attempt_decade_mean

decade=["1980~1990","1991~2000","2001~2010","2011~2018"]

decade_data=pd.DataFrame({"Year":decade,"three attempt mean":three_attempt_decade_mean,})
decade_data
```
